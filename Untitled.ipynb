```python
import os
import dask.dataframe as dd

# Ścieżka do danych
data_path = r"E:\marcel.furs\Documents\Jane-Data"
train_folder = os.path.join(data_path, "train.parquet")

# Wczytujemy wszystkie cechy (79 + target)
columns_to_load = ['date_id', 'time_id', 'symbol_id', 'responder_6'] + [f'feature_{i:02d}' for i in range(79)]

# **Wczytujemy Dask DataFrame, ale w małych blokach (~256MB zamiast całych 12GB)**
dask_train_data = dd.read_parquet(train_folder, columns=columns_to_load, blocksize="256MB")
```

```python
# Obliczamy braki danych
missing_data = dask_train_data.isnull().sum().compute(split_out=10)
total_rows = dask_train_data.shape[0].compute()
missing_percentage = (missing_data / total_rows) * 100

# Wyświetlenie wyników
print("Braki w danych (liczba braków):\n", missing_data)
print("\nBraki w danych (procentowo):\n", missing_percentage)

# Wizualizacja tylko TOP 20 cech
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
missing_percentage.nlargest(79).plot(kind='bar', color='purple')
plt.title("TOP cech z brakami danych")
plt.xlabel("Kolumny")
plt.ylabel("Procent braków")
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Pobranie próbki danych (10% dla oszczędności RAM)
sample_data = dask_train_data.sample(frac=0.1, random_state=42).compute()

# Obliczenie statystyk
mean_val = sample_data['responder_6'].mean()
median_val = sample_data['responder_6'].median()
std_val = sample_data['responder_6'].std()

# Rysowanie histogramu
plt.figure(figsize=(10, 6))
sns.histplot(sample_data['responder_6'], bins=30, kde=True, color='blue')

# Dodanie linii dla średniej i mediany
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Średnia: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {median_val:.2f}')

# Opcjonalnie: skala logarytmiczna
plt.yscale('log')  # Jeśli dane są mocno skośne, można to zakomentować

# Opisy wykresu
plt.title("Rozkład responder_6 (Próbka)")
plt.xlabel("Wartość responder_6")
plt.ylabel("Częstość (log)" if plt.yscale == 'log' else "Częstość")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pobranie próbki danych
sample_data = dask_train_data.sample(frac=0.1, random_state=42).compute()

# Pełna macierz korelacji
plt.figure(figsize=(12, 8))
sns.heatmap(sample_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Pełna macierz korelacji")
plt.show()

# Usunięcie cech z brakami powyżej 10%
missing_threshold = 0.10
missing_ratio = sample_data.isnull().mean()
filtered_features = missing_ratio[missing_ratio < missing_threshold].index
filtered_data = sample_data[filtered_features]

# Macierz korelacji po usunięciu cech z brakami
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Macierz korelacji po usunięciu cech z dużymi brakami")
plt.show()

# Wybór najbardziej skorelowanych cech
correlation_target = sample_data.corr()["responder_6"].abs().sort_values(ascending=False)
top_features = correlation_target.index[1:11]  # Top 10 najbardziej skorelowanych cech

# Macierz korelacji dla wybranych cech
plt.figure(figsize=(10, 6))
sns.heatmap(sample_data[top_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Macierz korelacji dla wybranych cech")
plt.show()
```

```python
# Reset indeksu, aby uniknąć duplikatów
sample_data = sample_data.reset_index(drop=True)

# Boxplot: Rozkład feature_01 względem przedziałów responder_6
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(sample_data['responder_6'], bins=5), y=sample_data['feature_01'])
plt.title("Rozkład feature_01 względem responder_6 (Próbka)")
plt.xlabel("Przedziały responder_6")
plt.ylabel("Wartość feature_01")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_regression

# Przygotowanie danych na próbce
X = sample_data[['feature_01', 'feature_02', 'feature_03']].fillna(0)
y = sample_data['responder_6']

# Obliczenie ważności cech
feature_scores = mutual_info_regression(X, y)
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': feature_scores})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# Wizualizacja ważności cech
plt.figure(figsize=(8, 5))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title("Ważność cech dla responder_6 (Próbka)")
plt.show()
```
