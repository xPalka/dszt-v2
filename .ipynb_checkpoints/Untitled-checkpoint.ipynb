```python
import os
import pyarrow as pa

# Ustawienie katalogu na dysku E dla plików tymczasowych
os.environ['TMPDIR'] = r"E:\Temp"
print("Lokalizacja plików tymczasowych:", os.environ['TMPDIR'])
```

```python
import pyarrow.parquet as pq
import pandas as pd

# Ścieżka do folderu z danymi
data_path = r"E:\marcel.furs\Documents\Jane-Data"
train_folder = os.path.join(data_path, "train.parquet")

# Przetwarzanie każdej partycji oddzielnie
all_data = []
columns_to_load = ['date_id', 'time_id', 'symbol_id', 'responder_6', 'feature_01', 'feature_02', 'feature_03']

for partition_id in range(10):  # Iterujemy przez partycje
    partition_path = os.path.join(train_folder, f"partition_id={partition_id}")
    print(f"Wczytywanie partycji {partition_id}...")
    
    # Wczytaj dane z ograniczonym zestawem kolumn
    partition_data = pq.read_table(partition_path, columns=columns_to_load).to_pandas()
    all_data.append(partition_data)

# Połączenie wszystkich danych w pamięci RAM
train_data = pd.concat(all_data, ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram responder_6
plt.figure(figsize=(10, 6))
sns.histplot(train_data['responder_6'], bins=30, kde=True, color='blue')
plt.title("Rozkład responder_6")
plt.xlabel("Wartość responder_6")
plt.ylabel("Częstość")
plt.savefig(r"E:\Temp\responder_6_histogram.png", dpi=300)
plt.show()

# Korelacja
corr_matrix = train_data[['responder_6', 'feature_01', 'feature_02', 'feature_03']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelacja cech z responder_6")
plt.savefig(r"E:\Temp\responder_6_correlation.png", dpi=300)
plt.show()
```

```python
import os

# Ustawienie katalogu tymczasowego na dysku E
os.environ['TMPDIR'] = r"E:\Temp"
print("Lokalizacja plików tymczasowych:", os.environ['TMPDIR'])
```

```python
import pyarrow.parquet as pq
import pandas as pd

# Ścieżka do folderu z danymi
data_path = r"E:\marcel.furs\Documents\Jane-Data"
train_folder = os.path.join(data_path, "train.parquet")

# Wczytanie wybranych kolumn
columns_to_load = ['date_id', 'time_id', 'symbol_id', 'responder_6', 'feature_01', 'feature_02', 'feature_03']

# Przetwarzanie danych partiami
all_data = []
for partition_id in range(10):  # Iterujemy przez partycje
    partition_path = os.path.join(train_folder, f"partition_id={partition_id}")
    print(f"Wczytywanie partycji {partition_id}...")
    
    # Wczytanie wybranych kolumn z partycji
    partition_data = pq.read_table(partition_path, columns=columns_to_load).to_pandas()
    all_data.append(partition_data)

# Łączenie danych w pamięci
train_data = pd.concat(all_data, ignore_index=True)
print("Rozmiar załadowanych danych:", train_data.shape)
```

```python
# Analiza braków danych
missing_data = train_data.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_data / len(train_data)) * 100
print("Braki w danych:\n", missing_percentage)

# Wizualizacja braków danych
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.heatmap(train_data.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Braki w danych")
plt.show()
```

```python
# Histogram responder_6
plt.figure(figsize=(10, 6))
sns.histplot(train_data['responder_6'], bins=30, kde=True, color='blue')
plt.title("Rozkład responder_6")
plt.xlabel("Wartość responder_6")
plt.ylabel("Częstość")
plt.show()
```

```python
# Macierz korelacji
corr_matrix = train_data[['responder_6', 'feature_01', 'feature_02', 'feature_03']].corr()

# Wizualizacja macierzy korelacji
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelacja cech z responder_6")
plt.show()
```
